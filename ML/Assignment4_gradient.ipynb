```python
currX = 2
learning_rate = 0.01
precision = 0.0000001
maxIteration = 1000
interation = 0
diff = 1
df = lambda x: 2*(x + 3)
```

```python
while diff > precision and interation < maxIteration:
  prevX = currX
  currX = prevX - (learning_rate * df(prevX))
  diff = abs(prevX - currX)
  interation = interation + 1;
  print("Iteration ",interation," x: ",currX)

print("Local minima : ", abs(currX)
```
